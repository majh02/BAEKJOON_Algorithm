```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier

import warnings
warnings.filterwarnings(action='ignore')
```

```python
direction = ['E', 'W', 'S', 'N']
```

### data 불러오기

```python
data = pd.read_excel('data.xlsx')
data
```

- X, y split

```python
target = 'y'
X = data.drop(target, axis=1)
y = data.loc[:, target].replace(['E','W','S','N'],[0,1,2,3]).values
```

```python
X.shape, y.shape
```

- train-test split

```python
X_train, y_train, X_test, y_test = X[:-4], y[:-4], X[-4:], y[-4:]
```

### ① Perceptron network

```python
# Perceptron network

clf = MLPClassifier(max_iter=300).fit(X_train, y_train) # training

# predict
pred_clf = clf.predict_proba(X_test)
for i, pred in enumerate(pred_clf):
    print(f'index {i} | answer: {direction[y_test[i]]}, predict: {direction[np.argmax(pred)]}')
```

### ② 2 Layer Perceptron (MLP)

```python
# 2 Layer Perceptron (MLP)

H = 2 # hidden layer = 2
clf2 = MLPClassifier(hidden_layer_sizes=(H,), learning_rate_init=0.1, max_iter=1000) # training
clf2.fit(X_train, y_train)

# predict
pred_clf2 = clf2.predict_proba(X_test)
for i, pred in enumerate(pred_clf2):
    print(f'index {i} | answer: {direction[y_test[i]]}, predict: {direction[np.argmax(pred)]}')
```
